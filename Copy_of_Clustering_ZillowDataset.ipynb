```python
# Code snippet 1
# Importing the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Zillow%20Tarrant%20County%20Raw%20Data.csv')

data
```

```python
# Code snippet 2
# Displaying the dataset columns
data.columns
```

```python
# Code snippet 3
# Listing all the categorical columns
cat_cols = ['property_type', 'exterior_walls', 'roof']

# Creating dummy columns/hot encoding for the categorical columns
encoded_data = pd.get_dummies(data, columns=cat_cols, drop_first = True, dtype = int)

# Display the encoded_data
encoded_data
```

```python
# Code snippet 4
# Importing K-means clustering algorithm
from sklearn.cluster import KMeans
```

```python
# Code snippet 5
# Assigning the algorithm as Kmeans
kmeans = KMeans(random_state  = 42)
# Code snippet 6
# Displaying the cluster group ids using Kmeans algo
data['Cluster Group'] = kmeans.fit_predict(encoded_data)

# Displaying the original dataset with the cluster group info
data
```

```python
# Code snippet 7
# Exploring the dataset with the cluster group ids
data.to_csv('Zillow Dataset with Clusters info.csv')
```

```python
# Code snippet 8
# Displaying the cluster groups and respective data points count
data['Cluster Group'].value_counts()
```
