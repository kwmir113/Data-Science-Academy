```python
# Code snippet 1
# Importing the income dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Tarrant%20County%20Income%20Dataset%20SQL%20CSV.csv')
data
```

```python
# Code snippet 2
# Importing sqlite3 package (to run SQL queries in Python)
import sqlite3
# Code snippet 3
# Creating a sample database as a connection
conn = sqlite3.connect('sample.db')
```

```python
# Code snippet 4
# Storing the pandas dataframe as a table in the sample database
data.to_sql('data_sql', conn)
```

```python
# Code snippet 5
# Display all the data from the table
sql_query = ''' SELECT * FROM data_sql '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 6
# Display all the data from the table where age > 50
sql_query = ''' SELECT * FROM data_sql WHERE age > 50 '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 6
# Display all the data from the table where age <= 50
sql_query = ''' SELECT * FROM data_sql WHERE age <=50 '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 7
# Display all the data from the table where age > 50
sql_query = ''' SELECT * FROM data_sql WHERE age > 50 '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 8
# Display all the data from the table where age < 30 and marital status of never married
sql_query = ''' SELECT * FROM data_sql WHERE age < 30 AND Marital_Status = 'Never-married'   '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 9
# Display all the data from the table where age < 30 and marital status as never married and with positive capital gains
sql_query = ''' SELECT * FROM data_sql WHERE age < 30 AND Marital_Status = 'Never-married' AND Capital_Gain > 0  '''
data_output = pd.read_sql(sql_query, conn)
data_output
```

```python
# Code snippet 9
# Display all the data from the table where age < 30 and marital status as never married and with positive capital gains
sql_query = ''' SELECT * FROM data_sql WHERE age < 30 AND Marital_Status = 'Never-married' AND Capital_Gain > 0  '''
data_output = pd.read_sql(sql_query, conn)
data_output
```
